```python
train=pd.read_csv('/kaggle/input/bike-sharing-demand/train.csv')
train['datetime']=train['datetime'].astype('datetime64')
train['time']=train['datetime'].dt.hour
train['year']=train['datetime'].dt.year
train['weekday']=train['datetime'].dt.weekday
```

```python
test=pd.read_csv('/kaggle/input/bike-sharing-demand/test.csv')
test['datetime']=test['datetime'].astype('datetime64')
test['time']=test['datetime'].dt.hour
test['year']=test['datetime'].dt.year
test['weekday']=test['datetime'].dt.weekday
test
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
plt.figure(figsize=(16,10))
sns.boxplot(train['temp'],train['count'],showfliers=False) 
```

```python
train.groupby('time')['count'].mean()
```

```python
train.groupby('temp')['count'].mean()
```

```python
train_2=train.drop(columns=['casual','registered','count','datetime'])
train_2
test_2=test.drop(columns=['datetime'])
test_2
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(train_2, train['count']) 
result = rf.predict(test_2)
result
```
